```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Capitalizacin total por sector ---
cap_sector = df.groupby("Sector")["Marketcap"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
cap_sector.plot(kind="bar", color="teal")
plt.title("Capitalizacin de Mercado por Sector (S&P 500)")
plt.ylabel("Capitalizacin total (USD)")
plt.xticks(rotation=45)
plt.show()


# --- Precio promedio por sector ---
price_sector = df.groupby("Sector")["Currentprice"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
price_sector.plot(kind="bar", color="orange")
plt.title("Precio Promedio de Acciones por Sector")
plt.ylabel("Precio promedio (USD)")
plt.xticks(rotation=45)
plt.show()


# --- Crecimiento promedio de ingresos por sector ---
growth_sector = df.groupby("Sector")["Revenuegrowth"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=growth_sector.index, y=growth_sector.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Crecimiento Promedio de Ingresos por Sector")
plt.ylabel("Revenue Growth (%)")
plt.show()

```

```python
import pandas as pd

# Ruta correcta
ruta_2 = "/content/drive/MyDrive/BOOTCAMP /DATOS/sp500_index.csv"
df_Index = pd.read_csv(ruta_2)   #  ahora s usamos ruta_2

# Revisar columnas
print(df_Index.columns)

# Ver primeras filas
df_Index.head()


```

```python
import matplotlib.pyplot as plt

# Asegurar tipo fecha
df_Index["Date"] = pd.to_datetime(df_Index["Date"])

# Calcular medias mviles
df_Index["SMA50"] = df_Index["S&P500"].rolling(window=50).mean()
df_Index["EMA200"] = df_Index["S&P500"].ewm(span=200, adjust=False).mean()

# Graficar
plt.figure(figsize=(14,6))
plt.plot(df_Index["Date"], df_Index["S&P500"], label="S&P 500", color="blue", alpha=0.6)
plt.plot(df_Index["Date"], df_Index["SMA50"], label="SMA 50", color="orange", linewidth=2)
plt.plot(df_Index["Date"], df_Index["EMA200"], label="EMA 200", color="red", linewidth=2)

plt.title("Resumen Ejecutivo de Tendencia del Mercado (SMA 50 vs EMA 200)")
plt.xlabel("Fecha")
plt.ylabel("Nivel del ndice")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

#**Resumen Ejecutivo de Tendencia del Mercado (SMA 50 vs. EMA 200)**


##Este anlisis de cruce de Medias Mviles determina la direccin y la fortaleza de la tendencia del S&P 500, proporcionando una visin clara para las decisiones de asignacin de capital a largo y mediano plazo.

Tendencia general:

1.El S&P 500 ha mantenido una trayectoria claramente alcista en la ltima dcada, con cadas puntuales (2018, 2020 por la pandemia, 2022 por ajuste monetario), pero recuperando con fuerza y alcanzando mximos en 2025.

2.Cruce de medias mviles (Golden/Death Cross):

-Cuando la SMA 50 cruza por encima de la EMA 200, se genera una seal de fortaleza (Golden Cross), anticipando subidas sostenidas.

-Cuando la SMA 50 cae por debajo de la EMA 200, se interpreta como debilidad (Death Cross), alertando sobre correcciones o mercados bajistas.

3.Observaciones recientes:

-Entre 2022 y 2023 hubo un cruce bajista (Death Cross), reflejando la cada del mercado en ese perodo.

-A partir de 2023-2024 se observa un nuevo cruce alcista (Golden Cross), confirmando la recuperacin y fortalecimiento de la tendencia.

Actualmente (2025), la SMA 50 se mantiene por encima de la EMA 200, lo que respalda un escenario de continuidad alcista en el mediano plazo.

```python
import matplotlib.pyplot as plt

# Usamos el dataframe df (sp500_companies.csv)
# Proxy de Retorno: Revenuegrowth promedio por sector
# Proxy de Riesgo: desviacin estndar de Marketcap por sector

risk_return = df.groupby("Sector").agg({
    "Revenuegrowth": "mean",   # rendimiento medio
    "Marketcap": "std"         # dispersin como proxy de riesgo
}).rename(columns={"Revenuegrowth": "Retorno", "Marketcap": "Riesgo"})

# Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(risk_return["Riesgo"], risk_return["Retorno"], s=150, alpha=0.7, color="teal")

for sector, row in risk_return.iterrows():
    plt.annotate(sector, (row["Riesgo"], row["Retorno"]))

plt.title("Mapa RiesgoRetorno Sectorial (S&P 500)")
plt.xlabel("Riesgo (Std Marketcap)")
plt.ylabel("Retorno (Revenue Growth medio)")
plt.grid(True)
plt.show()

```

#**Mapa Estratgico de RiesgoRetorno Sectorial (S&P 500)**

El grfico muestra la relacin entre riesgo (volatilidad en capitalizacin de mercado, eje X) y retorno (crecimiento promedio de ingresos, eje Y) para los principales sectores del S&P 500.

-Technology y Communication Services destacan por su alto crecimiento, pero con mayor riesgo.

-Financial Services y Real Estate ofrecen una relacin ms equilibrada, combinando retornos slidos con riesgos moderados.

-Energy y Consumer Defensive presentan bajo crecimiento, pero funcionan como sectores defensivos y estables en periodos de incertidumbre.

En conclusin, el mapa permite identificar qu sectores son ms atractivos segn el apetito de riesgo: los dinmicos (alto retornoalto riesgo), los defensivos (bajo retornobajo riesgo) y los estratgicamente balanceados.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ejemplo de dataframe (puedes reemplazarlo por el tuyo)
data = {
    "Sector": ["Technology", "Healthcare", "Financials", "Energy", "Consumer Defensive"],
    "PE": [28, 22, 16, 10, 18],
    "HistoricalAvg_PE": [25, 20, 15, 12, 17],
    "EV_EBITDA": [15, 13, 9, 6, 10],
    "HistoricalAvg_EVEBITDA": [14, 12, 10, 7, 9]
}

df_val = pd.DataFrame(data)

# Crear subplots
fig, ax = plt.subplots(1, 2, figsize=(14,6))

# Grfico P/E
df_val.plot(x="Sector", y=["PE","HistoricalAvg_PE"], kind="bar", ax=ax[0], color=["#ff7f0e","#1f77b4"])
ax[0].set_title("Valoracin Relativa por Sector - P/E")
ax[0].set_ylabel("Ratio P/E")
ax[0].legend(["P/E Actual","Promedio Histrico"])

# Grfico EV/EBITDA
df_val.plot(x="Sector", y=["EV_EBITDA","HistoricalAvg_EVEBITDA"], kind="bar", ax=ax[1], color=["#2ca02c","#9467bd"])
ax[1].set_title("Valoracin Relativa por Sector - EV/EBITDA")
ax[1].set_ylabel("EV/EBITDA")
ax[1].legend(["EV/EBITDA Actual","Promedio Histrico"])

plt.suptitle("Comparacin de Valoracin Relativa Sectorial (S&P 500)", fontsize=14)
plt.tight_layout()
plt.show()
```

#**Valoracin Relativa Sectorial (P/E y EV/EBITDA  S&P 500)**

Las mtricas de valoracin permiten identificar si un sector est caro o barato en comparacin con su propio promedio histrico o con el mercado en general.

##**P/E (Precio/Utilidad):**

Alto  el mercado paga ms por cada dlar de ganancia, lo que refleja expectativas de crecimiento, pero tambin un posible riesgo de sobrevaloracin.

Bajo  puede indicar que el sector est infravalorado o que enfrenta problemas de rentabilidad.

###**V/EBITDA (Valor Empresa / EBITDA):**

Es una mtrica ms completa porque considera deuda y capital.

Alto  sector relativamente caro, con expectativas de retorno alto pero riesgo de correccin.

Bajo  sector barato en trminos relativos, potencialmente atractivo para inversin de valor.

####**Cmo se interpreta en el mapa:**

Sectores por encima de su promedio histrico  posibles oportunidades de venta o sectores calientes.

Sectores por debajo del promedio  posibles oportunidades de compra (mercado descontado).
