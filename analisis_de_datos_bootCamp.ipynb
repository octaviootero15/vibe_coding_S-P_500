```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Capitalización total por sector ---
cap_sector = df.groupby("Sector")["Marketcap"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
cap_sector.plot(kind="bar", color="teal")
plt.title("Capitalización de Mercado por Sector (S&P 500)")
plt.ylabel("Capitalización total (USD)")
plt.xticks(rotation=45)
plt.show()


# --- Precio promedio por sector ---
price_sector = df.groupby("Sector")["Currentprice"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
price_sector.plot(kind="bar", color="orange")
plt.title("Precio Promedio de Acciones por Sector")
plt.ylabel("Precio promedio (USD)")
plt.xticks(rotation=45)
plt.show()


# --- Crecimiento promedio de ingresos por sector ---
growth_sector = df.groupby("Sector")["Revenuegrowth"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=growth_sector.index, y=growth_sector.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Crecimiento Promedio de Ingresos por Sector")
plt.ylabel("Revenue Growth (%)")
plt.show()

```

```python
import matplotlib.pyplot as plt

# Asegurar tipo fecha
df_Index["Date"] = pd.to_datetime(df_Index["Date"])

# Calcular medias móviles
df_Index["SMA50"] = df_Index["S&P500"].rolling(window=50).mean()
df_Index["EMA200"] = df_Index["S&P500"].ewm(span=200, adjust=False).mean()

# Graficar
plt.figure(figsize=(14,6))
plt.plot(df_Index["Date"], df_Index["S&P500"], label="S&P 500", color="blue", alpha=0.6)
plt.plot(df_Index["Date"], df_Index["SMA50"], label="SMA 50", color="orange", linewidth=2)
plt.plot(df_Index["Date"], df_Index["EMA200"], label="EMA 200", color="red", linewidth=2)

plt.title("Resumen Ejecutivo de Tendencia del Mercado (SMA 50 vs EMA 200)")
plt.xlabel("Fecha")
plt.ylabel("Nivel del Índice")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

#**Resumen Ejecutivo de Tendencia del Mercado (SMA 50 vs. EMA 200)**


##Este análisis de cruce de Medias Móviles determina la dirección y la fortaleza de la tendencia del S&P 500, proporcionando una visión clara para las decisiones de asignación de capital a largo y mediano plazo.

Tendencia general:

1.El S&P 500 ha mantenido una trayectoria claramente alcista en la última década, con caídas puntuales (2018, 2020 por la pandemia, 2022 por ajuste monetario), pero recuperando con fuerza y alcanzando máximos en 2025.

2.Cruce de medias móviles (Golden/Death Cross):

-Cuando la SMA 50 cruza por encima de la EMA 200, se genera una señal de fortaleza (“Golden Cross”), anticipando subidas sostenidas.

-Cuando la SMA 50 cae por debajo de la EMA 200, se interpreta como debilidad (“Death Cross”), alertando sobre correcciones o mercados bajistas.

3.Observaciones recientes:

-Entre 2022 y 2023 hubo un cruce bajista (Death Cross), reflejando la caída del mercado en ese período.

-A partir de 2023-2024 se observa un nuevo cruce alcista (Golden Cross), confirmando la recuperación y fortalecimiento de la tendencia.

Actualmente (2025), la SMA 50 se mantiene por encima de la EMA 200, lo que respalda un escenario de continuidad alcista en el mediano plazo.

```python
import matplotlib.pyplot as plt

# Usamos el dataframe df (sp500_companies.csv)
# Proxy de Retorno: Revenuegrowth promedio por sector
# Proxy de Riesgo: desviación estándar de Marketcap por sector

risk_return = df.groupby("Sector").agg({
    "Revenuegrowth": "mean",   # rendimiento medio
    "Marketcap": "std"         # dispersión como proxy de riesgo
}).rename(columns={"Revenuegrowth": "Retorno", "Marketcap": "Riesgo"})

# Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(risk_return["Riesgo"], risk_return["Retorno"], s=150, alpha=0.7, color="teal")

for sector, row in risk_return.iterrows():
    plt.annotate(sector, (row["Riesgo"], row["Retorno"]))

plt.title("Mapa Riesgo–Retorno Sectorial (S&P 500)")
plt.xlabel("Riesgo (Std Marketcap)")
plt.ylabel("Retorno (Revenue Growth medio)")
plt.grid(True)
plt.show()

```

#**Mapa Estratégico de Riesgo–Retorno Sectorial (S&P 500)**

El gráfico muestra la relación entre riesgo (volatilidad en capitalización de mercado, eje X) y retorno (crecimiento promedio de ingresos, eje Y) para los principales sectores del S&P 500.

-Technology y Communication Services destacan por su alto crecimiento, pero con mayor riesgo.

-Financial Services y Real Estate ofrecen una relación más equilibrada, combinando retornos sólidos con riesgos moderados.

-Energy y Consumer Defensive presentan bajo crecimiento, pero funcionan como sectores defensivos y estables en periodos de incertidumbre.

En conclusión, el mapa permite identificar qué sectores son más atractivos según el apetito de riesgo: los dinámicos (alto retorno–alto riesgo), los defensivos (bajo retorno–bajo riesgo) y los estratégicamente balanceados.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ejemplo de dataframe (puedes reemplazarlo por el tuyo)
data = {
    "Sector": ["Technology", "Healthcare", "Financials", "Energy", "Consumer Defensive"],
    "PE": [28, 22, 16, 10, 18],
    "HistoricalAvg_PE": [25, 20, 15, 12, 17],
    "EV_EBITDA": [15, 13, 9, 6, 10],
    "HistoricalAvg_EVEBITDA": [14, 12, 10, 7, 9]
}

df_val = pd.DataFrame(data)

# Crear subplots
fig, ax = plt.subplots(1, 2, figsize=(14,6))

# Gráfico P/E
df_val.plot(x="Sector", y=["PE","HistoricalAvg_PE"], kind="bar", ax=ax[0], color=["#ff7f0e","#1f77b4"])
ax[0].set_title("Valoración Relativa por Sector - P/E")
ax[0].set_ylabel("Ratio P/E")
ax[0].legend(["P/E Actual","Promedio Histórico"])

# Gráfico EV/EBITDA
df_val.plot(x="Sector", y=["EV_EBITDA","HistoricalAvg_EVEBITDA"], kind="bar", ax=ax[1], color=["#2ca02c","#9467bd"])
ax[1].set_title("Valoración Relativa por Sector - EV/EBITDA")
ax[1].set_ylabel("EV/EBITDA")
ax[1].legend(["EV/EBITDA Actual","Promedio Histórico"])

plt.suptitle("Comparación de Valoración Relativa Sectorial (S&P 500)", fontsize=14)
plt.tight_layout()
plt.show()
```

#**Valoración Relativa Sectorial (P/E y EV/EBITDA – S&P 500)**

Las métricas de valoración permiten identificar si un sector está “caro” o “barato” en comparación con su propio promedio histórico o con el mercado en general.

##**P/E (Precio/Utilidad):**

Alto → el mercado paga más por cada dólar de ganancia, lo que refleja expectativas de crecimiento, pero también un posible riesgo de sobrevaloración.

Bajo → puede indicar que el sector está infravalorado o que enfrenta problemas de rentabilidad.

###**V/EBITDA (Valor Empresa / EBITDA):**

Es una métrica más completa porque considera deuda y capital.

Alto → sector relativamente caro, con expectativas de retorno alto pero riesgo de corrección.

Bajo → sector barato en términos relativos, potencialmente atractivo para inversión de valor.

####**Cómo se interpreta en el mapa:**

Sectores por encima de su promedio histórico → posibles oportunidades de venta o sectores “calientes”.

Sectores por debajo del promedio → posibles oportunidades de compra (mercado descontado).
